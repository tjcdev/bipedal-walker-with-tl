```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Baseline

#### Bipedal Walker (base-1)

```python
base_bip = pd.read_json('base-bip-10m.json')
base_bip['code'] = 'base-1'
sns_plot = sns.lineplot(x="nupdates", y="eprewmean",
             ci="explained_variance",
             err_style="band",
             data=base_bip)
plt.ylim(-160, 300)
plt.show()
sns_plot.get_figure().savefig("baseline_normal.png",  bbox_inches = 'tight')
```

#### Biphard (base-2)

```python
base_biphard = pd.read_json('base-biphard-10m.json')
base_biphard['code'] = 'base-2'
sns_plot = sns.lineplot(x="nupdates", y="eprewmean", data=base_biphard)
plt.ylim(-160, 50)
plt.show()

sns_plot.get_figure().savefig("baseline_hardcore.png",  bbox_inches = 'tight')
```

# Dillation(?)

```python
dill = pd.read_json('dill.json')
dill['code'] = 'dill'
sns_plot = sns.lineplot(x="nupdates", y="eprewmean",
             ci="explained_variance",
             err_style="band",
             data=dill)
plt.ylim(-160, 200)
plt.show()

sns_plot.get_figure().savefig("dilation.png",  bbox_inches = 'tight')
```

#### Bip with Dillation

```python
dill_bip = dill.iloc[::2, :]
sns_plot = sns.lineplot(x="nupdates", y="eprewmean",
             ci="explained_variance",
             err_style="band",
             data=dill_bip)
plt.ylim(-160, 200)
plt.show()
```

#### Biphard with Dill

```python
dill_biphard = dill.iloc[1::2, :]
sns_plot = sns.lineplot(x="nupdates", y="eprewmean",
             ci="explained_variance",
             err_style="band",
             data=dill_biphard)
plt.ylim(-160, 200)
plt.show()
```
